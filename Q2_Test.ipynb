```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
spQ2 = pd.read_csv("spQ2.csv")
spQ2.head()
```

```python
Target = pd.DataFrame({'Target_status':[1,2,3], 'Target': ['Dropout', 'Graduate', 'Enrolled'] })
spQ2Lable = spQ2.merge(Target, on='Target_status')
```

```python

plt.figure(figsize=(6,4))
sns.countplot(x="Target_status", data=spQ2, order=spQ2["Target_status"].value_counts().index)
plt.title("Distribution of Student Status")
plt.xlabel("Student Status")
plt.ylabel("Count")
plt.show()
```

```python
categorical_cols = spQ2[["Mother\'s qualification", "Father\'s qualification", 
                    "Mother\'s occupation", "Father's occupation"]]

for col in categorical_cols:
    print(f"\n--- {col} ---")
    print(spQ2[col].value_counts())   
    print(spQ2[col].value_counts(normalize=True))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(y=col, data=spQ2, order=spQ2[col].value_counts().index)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
continuous_cols = spQ2[["GDP", "Inflation rate", "Unemployment rate"]]
for col in continuous_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=spQ2[col], color="lightgreen")
    plt.title(f"Boxplot of {col}")
    plt.show()
```
