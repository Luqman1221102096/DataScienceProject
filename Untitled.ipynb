```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
spQ2 = pd.read_csv("spQ2.csv")
spQ2.head()
```

```python
Target_Status = pd.DataFrame({'Target':['Dropout', 'Graduate', 'Enrolled'], 'Target_status': [1,2,3]})
clean_sp = spQ2.merge(Target_Status, on='Target')
num_sp = clean_sp.drop(columns=['Target'])
num_sp.head()
```

```python
sns.heatmap(
    num_sp.corr()[["Target_status"]],  # correlation only for one column
    cmap="cividis",
    annot=True
)
```

```python
sns.heatmap(
    spQ2.corr()[["Target_status"]],  # correlation only for one column
    cmap="cividis",
    annot=True
)
```
